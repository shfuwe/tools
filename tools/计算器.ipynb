```python
math.log(100, 10)
```

```python
def xn(n):
    if n==0:
        return 0
    return n*math.log(n, 2)
```

```python
xn(4/6)+xn(2/6)
```

```python
def cc(n):
    return -(xn(n)+xn(1-n))
```

```python
cc(1/6)
```

```python
cc(1)
```
